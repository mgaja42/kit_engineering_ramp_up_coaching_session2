```python
import scipy.stats as stats 
import numpy as np
import matplotlib.pyplot as plt
```

```python
def calculate_p_value(n1: int, n2: int, N: int) -> float:
    #generating N coin tosses with probability of heads equal to n1
    sample1 = np.random.binomial(1, n1, N)
    #generating N coin tosses with probability of heads equal to n2
    sample2 = np.random.binomial(1, n2, N)
    
    p_value = stats.ttest_ind(a=sample1, b=sample2)[1]
    return p_value
```

```python
p_value = [calculate_p_value(0.5, 0.45, i) for i in range(1000)]
```

```python
plt.plot(p_value)
plt.title("p_values for two coins")
plt.xlabel("Number of Tosses")
plt.ylabel("p_value")
```
